```python
from collections import deque

N, M, V = map(int, input().split()) # 정점의 개수, 간선의 개수, 시작점
graph = [[] for _ in range(N+1)] # n+1개의 빈리스트들을 리스트안에 넣어줌, n+1인 이유는 안 헷갈리기 위해
for _ in range(M): # 간선의 개수만큼 돈다돌아
    a, b = map(int, input().split()) # 간선 이어져있는 상태 입력
    graph[a].append(b)
    graph[a].sort()
    graph[b].append(a)
    graph[b].sort() # 이 4줄에선 정점 인덱스 별로 이어진 정점이 입력됨 ex) [[], [2, 3, 4], [1, 4], [1, 4], [1, 2, 3]]


def dfs(graph, v, visited): # 깊이 우선 탐색
    visited[v] = True # v로 들어가는, 한번 돌때마다의 시작점을 True = 방문함!으로 바꿔줌
    print(v, end=' ') # v를 출력 (방문한 정점)
    for i in graph[v]: # 그래프의 v번째를 i가 받아용, ex) 첫번째면 2, 3, 4 <-- 이 for문은 2 한번 돌고 3 한번 돌고 이럼...
        if not visited[i]: # 방문을 안했다는 뜻인가?, ex) 2, 3, 4 다 방문을 안했으면?
            dfs(graph, i, visited) # 재귀함수가 실행됨, 아마 2가 들어가겠지?

# 2가 들어가서 다시 실행되면 2가 출력되고 // 그래프의 2번째를 받아옴 --> 1, 4 // 근데 1은 시작할때 True로 바뀜 // 
# for문이 그럼 1, 4니까 4를 받아옴 // 4로 다시 실행.... 반복

visited = [False] * (N+1)
dfs(graph, V, visited) # 이어진 상태, 시작점, n+1만큼 False가 들어있는 리스트 <-- 중요
print()

def bfs(graph, start, visited):
    queue = deque([start]) # 시작점, 이건 리스트가 아니라 deque로 넣어줌, deque가 뭘까...?
    visited[start] = True # 돌때마다 현재 위치를 True로 바꿔줌
    while queue: # queue가 존재하는 한 반복
        v = queue.popleft() # queue의 왼쪽부터 v에 넣고 queue에서는 빼버림
        print(v, end= ' ') # 현재 위치 출력
        for i in graph[v]: # 현재 이어진 모든 상태에서 현재위치번째를 가져옴
            if not visited[i]: # 아까랑 마찬가지로 2, 3, 4가 들어갈텐데...
                queue.append(i) # 2가 False니까 queue에는 2가 들어감
                visited[i] = True # 2는 방문, True로 바뀜
                
# 시작점이 deque로 들어감 --> queue에서 팝!하고 빠짐, 그게 현재위치가 되고 출력됨

visited = [False] * (N+1)
bfs(graph, V, visited)


# 4 5 1
# 1 2
# 1 3
# 1 4
# 2 4
# 3 4

# dfs : 1 2 4 3 
# bfs : 1 2 3 4 

#####################################

# 1000 1 1000
# 999 1000

# 5 5 3
# 5 4
# 5 2
# 1 2
# 3 4
# 3 1
```

```python
from collections import deque
n = int(input())
m = int(input())
graph = [[] for _ in range(n+1)]
for _ in range(m):
  a, b = map(int, input().split())
  graph[a].append(b)
  graph[b].append(a)



q = deque()
q.append(1)
visited = [False] * (n+1)
visited[1] = True
while q:
  node = q.popleft()
  for i in graph[node]:
    if not visited[i]:
      visited[i] = True
      q.append(i)



print(visited.count(True) - 1)
```

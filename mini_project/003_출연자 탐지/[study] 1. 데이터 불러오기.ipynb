* 위에서 저장한 "video_list"에 있는 파일들을 분할하여 이미지 데이터로 만드세요.
* 이미지 저장 경로 : IMAGE_PATH = ROOT_PATH + "/image" <br>

> 파일의 경로가 저장된 video_list를 for문을 활용해서 하나씩 불러오세요.

```python
# 실습해보세요.

IMAGE_PATH = 'C:\\Users\\lak50\\Desktop\\aivle\\Aivle_Lecture\\028_mini_project\\practice\\AIVLE3rd_individual\\image1'
for i in range(4):
  name_list[i] = cv2.VideoCapture(path_list[i])
print(name_list)
name_str_list = ['sayuri', 'sam', 'park', 'jung']

TIME_MEASUERMENT_UNIT = 1

for i in range (4):
  while name_list[i].isOpened():
    ret,frame = name_list[i].read()
    if ret:
    # 현재 프레임 위치 (msec) 
      frame_sec = name_list[i].get(cv2.CAP_PROP_POS_MSEC)/1000
      if frame_sec.is_integer():
        if (frame_sec % TIME_MEASUERMENT_UNIT == 0):
          filename = IMAGE_PATH + "/" + name_str_list[i] + "/" + str(round(frame_sec)) + ".jpg"
          print(filename)
          cv2.imwrite(filename, frame) 
    else:
      break

name_list[i].release()
#   while name_list[i].isOpened():
#     ret, frame = name_list[i].read()
#     if ret:
#       frame_sec = name_list[i].get(cv2.CAP_PROP_POS_MSEC)/1000
#       if frame_sec.is_integer():
#         if (frame_sec % TIME_MEASUERMENT_UNIT == 0):
#           filename = IMAGE_PATH + "/" + name_str_list[i] + "/" + str(round(frame_sec)) + ".jpg"
#           cv2.imwrite(filename, frame)
#   # else:
#   #   break
#   # filename = IMAGE_PATH + "/" + name_str_list[i] + "/" + str(count) + ".jpg"
#   # cv2.imwrite(filename, frame)
#     else:
#       break
#   name_list[i].release()
#     # if i > 3:
#     #   break
#     # if int(name_list[i].get(cv2.CAP_PROP_FRAME_COUNT)) == count:
#     #   i += 1



```

```python
# 값이 하나밖에 없음
# 지역 : 서울 종로구, 망 : 도시대기, 측정소명 : 종로구, 주소 : 서울 종로구 종로35가길 19, 측정소코드 : 111123
drop_col_air = ['지역', '망', '측정소명', '주소', '측정소코드', '측정일시']
air_21_time = air_21['측정일시']
air_22_time = air_22['측정일시']
air_21_drop = air_21.drop(drop_col_air, axis = 1,inplace = False)
air_22_drop = air_22.drop(drop_col_air, axis = 1,inplace = False)
drop_col_weather = ['기온 QC플래그','해면기압 QC플래그','지면상태(지면상태코드)','풍속 QC플래그','풍향 QC플래그','습도 QC플래그', '현지기압 QC플래그', '일시']
weather_21_time = weather_21['일시']
weather_22_time = weather_22['일시']
weather_21_drop = weather_21.drop(drop_col_weather, axis = 1,inplace = False)
weather_22_drop = weather_22.drop(drop_col_weather, axis = 1,inplace = False)
print(weather_22_drop.columns, weather_21_drop.info())
print('=='*40)
print(weather_21_drop.columns, weather_22_drop.info())
print('=='*40)
print(air_21_drop.columns, air_21_drop.info())
print('=='*40)
print(air_22_drop.columns, air_22_drop.info())
```

qc 플래그들은 0 또는 9이므로 -1로 채움 --> 자료가 정확한지 평가하는 기준인 것 같음  
지면온도(°C), 5cm 지중온도(°C), 10cm 지중온도(°C), 20cm 지중온도(°C), 30cm 지중온도(°C), 풍속(m/s)은 전부다 interpolate로

```python
display(weather_21_drop.isna().sum())
display(weather_22_drop.isna().sum())
display(weather_22_drop.info())
display(air_21_drop.info())
```

```python
from sklearn.impute import KNNImputer
# 아래에 실습코드를 작성하고 결과를 확인합니다.
pd.set_option('display.max_columns', 40)
pd.set_option('display.max_row', 40)
weather_21_drop.fillna({'강수량 QC플래그': -1, '일조 QC플래그': -1, '일사 QC플래그' : -1, '지면온도 QC플래그' : -1}, inplace = True)
weather_22_drop.fillna({'강수량 QC플래그': -1, '일조 QC플래그': -1, '일사 QC플래그' : -1, '지면온도 QC플래그' : -1}, inplace = True)
for i in range(23, 29):
    weather_21_interp = weather_21_drop.iloc[:, i]
    weather_22_interp = weather_22_drop.iloc[:, i]
    weather_21_interp.interpolate(method = 'linear', inplace = True)
    weather_22_interp.interpolate(method = 'linear', inplace = True)
    weather_21_drop.iloc[:, i] = weather_21_interp
    weather_22_drop.iloc[:, i] = weather_22_interp
# 위는 qc와 interpolate
weather_21 = weather_21_drop.dropna(subset=['풍속(m/s)', '풍향(16방위)', '전운량(10분위)'])
weather_22 = weather_22_drop.dropna(subset=['풍속(m/s)', '풍향(16방위)', '전운량(10분위)'])

dumm_var = ['지점명', '운형(운형약어)']
weather_21_dum = pd.get_dummies(weather_21, columns = dumm_var, drop_first = True)
weather_22_dum = pd.get_dummies(weather_22, columns = dumm_var, drop_first = True)
# display(weather_21_dum)
# display(weather_21_dum.columns)
# 임퓨터 선언(4개의 평균으로 계산하겠다)
imputer=KNNImputer(n_neighbors=4)

#임퓨터를 사용하여 filled_train으로 저장 이후 같은 임퓨터를 사용할때는 imputer.transform()으로 사용하면됨
weather_21_KNN = imputer.fit_transform(weather_21_dum)
weather_22_KNN = imputer.fit_transform(weather_22_dum)
air_21_KNN = imputer.fit_transform(air_21_drop)
air_22_KNN = imputer.fit_transform(air_22_drop)
#사용하면 array값으로 나오기때문에 dataframe으로 바꿔주고 컬럼을가져옴
weather_21_KNN = pd.DataFrame(weather_21_KNN, columns=weather_21_dum.columns)
weather_22_KNN = pd.DataFrame(weather_22_KNN, columns=weather_22_dum.columns)
air_21_KNN = pd.DataFrame(air_21_KNN, columns=air_21_drop.columns)
air_22_KNN = pd.DataFrame(air_22_KNN, columns=air_22_drop.columns)
```

- weather_21_KNN, weather_22_KNN, air_21_KNN, air_22_KNN가 최종 결측치 처리 데이터
- weather_21_time, weather_22_time, air_21_time, air_22_time가 시간순 열, KNN돌리느라 잠깐 빼놓음

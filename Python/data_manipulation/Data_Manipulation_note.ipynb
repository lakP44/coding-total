# Rolling, Shift 하는법

```python
stock['Close'].rolling(3).mean() # 쓰면 0~2행 값의 평균이 2행에 저장, 1~3행 평균에 3행저장...

stock['close'].shift(1) # 쓰면 1행씩 아래로 밀림, 음수는 땡겨오기

stock['close'].rolling(3, min_periods = 1).mean() # 을 쓰면 0행값은 그대로 0행, 0~1행의 평균은 1행, 그 다음은 그대로 감 --> NaN값이 없음
```

```python
titanic.pibot_table('cnt', 'Sex', 'Survived', aggfunc = np.sum, fill_value = 1, margins = True, margins_name = '분석결과')
# aggfunc는 집계함수, fill_value는 NaN값 채우기, margins는 분석결과를 오른쪽과 아래에 붙일지 여부, margins_name는 마진열의 이름
```
